```python
import pandas as pd
import numpy as np

np.random.seed(42)
```

Generating mock_users.csv

```python
num_users = 200

roles = ['HR', 'Recruiter', 'Founder', 'Analyst']
plans = ['free', 'trial', 'paid']
channels = ['LinkedIn', 'Referral', 'Website', 'Event']

user_data = {
    'user_id': [f'U{i:04d}' for i in range(1, num_users+1)],
    'email': [f'user{i}@example.com' for i in range(1, num_users+1)],
    'signup_timestamp': pd.to_datetime('2023-01-01') + pd.to_timedelta(np.random.randint(0, 180, num_users), unit='D'),
    'role': np.random.choice(roles, num_users),
    'plan_type': np.random.choice(plans, num_users, p=[0.5, 0.3, 0.2]),
    'acquisition_channel': np.random.choice(channels, num_users),
    'country': np.random.choice(['US', 'India', 'UK', 'Germany'], num_users)
}

df_users = pd.DataFrame(user_data)

df_users.to_csv('mock_users.csv', index=False)
```

Generating mock_sessions.csv

```python
num_sessions = 500

session_data = {
    'session_id': [f'S{i:05d}' for i in range(1, num_sessions+1)],
    'user_id': np.random.choice(df_users['user_id'], num_sessions),
    'session_start': pd.to_datetime('2023-01-01') + pd.to_timedelta(np.random.randint(0, 180, num_sessions), unit='D'),
    'session_end': pd.to_datetime('2023-01-01') + pd.to_timedelta(np.random.randint(0, 180, num_sessions), unit='D'),
    'device_type': np.random.choice(['desktop', 'mobile', 'tablet'], num_sessions, p=[0.6, 0.3, 0.1]),
    'location': np.random.choice(['US', 'India', 'UK', 'Germany'], num_sessions)
}

df_sessions = pd.DataFrame(session_data)

df_sessions['session_end'] = df_sessions.apply(
    lambda row: row['session_end'] if row['session_end'] >= row['session_start'] else row['session_start'], axis=1
)

df_sessions.to_csv('mock_sessions.csv', index=False)
```

Generating mock_events.csv

```python
num_events = 1000

event_names = ['signup', 'login', 'view_heatmap', 'search_role', 'apply_filter', 'view_company', 'logout']
feature_areas = ['Dashboard', 'Search', 'Insights', 'Company']

event_data = {
    'event_id': [f'E{i:05d}' for i in range(1, num_events+1)],
    'user_id': np.random.choice(df_users['user_id'], num_events),
    'session_id': np.random.choice(df_sessions['session_id'], num_events),
    'event_name': np.random.choice(event_names, num_events, p=[0.05, 0.15, 0.3, 0.2, 0.15, 0.1, 0.05]),
    'event_timestamp': pd.to_datetime('2023-01-01') + pd.to_timedelta(np.random.randint(0, 180, num_events), unit='D'),
    'feature_area': np.random.choice(feature_areas, num_events),
    'metadata': ['{}' for _ in range(num_events)]
}

df_events = pd.DataFrame(event_data)

df_events.to_csv('mock_events.csv', index=False)
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/mock_users.csv')

print("Missing values before cleaning:")
print(df.isna().sum())

for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna('Unknown')
    else:
        non_null_values = df[col].dropna()
        if len(non_null_values) > 0:
            min_val, max_val = non_null_values.min(), non_null_values.max()
            random_fill = np.random.uniform(min_val, max_val, size=df[col].isna().sum())
            df.loc[df[col].isna(), col] = random_fill
        else:
            df[col] = np.random.uniform(0, 1, size=len(df))

df = df.round(2)

df.to_csv('/content/mock_users_cleaned.csv', index=False)

print("\nCleaned file saved as mock_users_cleaned.csv")
print("Missing values after cleaning:")
print(df.isna().sum())

df.head()
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/mock_users.csv')

date_cols = ['last_login_date', 'signup_date']
for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')

for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].fillna('Unknown')

for col in df.select_dtypes(include=[np.number]).columns:
    non_null = df[col].dropna()
    if len(non_null) > 0:
        df.loc[df[col].isna(), col] = np.random.uniform(non_null.min(), non_null.max(), size=df[col].isna().sum())
    else:
        df[col] = np.random.uniform(0, 1, size=len(df))

if 'last_login_date' in df.columns:
    valid_dates = df['last_login_date'].dropna()
    min_date, max_date = valid_dates.min(), valid_dates.max()

    n_missing = df['last_login_date'].isna().sum()
    random_days = np.random.randint(0, (max_date - min_date).days + 1, n_missing)
    random_dates = [min_date + pd.Timedelta(days=int(d)) for d in random_days]

    df.loc[df['last_login_date'].isna(), 'last_login_date'] = random_dates

if all(col in df.columns for col in ['signup_date', 'last_login_date']):
    df.loc[df['signup_date'] > df['last_login_date'], 'last_login_date'] = df['signup_date'] + pd.to_timedelta(np.random.randint(1, 30, size=len(df)), unit='D')

df.to_csv('/content/mock_users_cleaned.csv', index=False)

print("Cleaned file saved as mock_users_cleaned.csv")
print("Missing values after cleaning:")
print(df.isna().sum())
df.head()
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/mock_users.csv')

df['signup_date'] = pd.to_datetime(df['signup_date'], errors='coerce')

valid_signup_dates = df['signup_date'].dropna()

if len(valid_signup_dates) > 0:
    min_date = valid_signup_dates.min()
    max_date = valid_signup_dates.max()

    n_missing = df['signup_date'].isna().sum()

    random_offsets = np.random.randint(0, (max_date - min_date).days + 1, n_missing)

    random_signup_dates = [min_date + pd.Timedelta(days=int(offset)) for offset in random_offsets]
    df.loc[df['signup_date'].isna(), 'signup_date'] = random_signup_dates

else:
    df['signup_date'] = pd.to_datetime(
        np.random.choice(pd.date_range('2023-01-01', '2023-12-31'), size=len(df))
    )

print("Missing signup dates filled successfully!")
print(df['signup_date'].describe())

df.to_csv('/content/mock_users_cleaned.csv', index=False)
print("Cleaned file saved as mock_users_cleaned.csv")
```
